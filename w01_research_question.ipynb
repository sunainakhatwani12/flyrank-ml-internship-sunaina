## 2. The question: decision, action, cost of a wrong call

### Research question

**Which pages should a content editor review first because they show enough demand and evidence of decline, staleness, weak CTR, or weak engagement to justify attention?**

The **unit of analysis** is one pseudonymized content page. The intended **output** is a ranked review queue containing a score, a suggested action, a reason code, and a confidence level.

The decision this work should improve is how an editor uses limited review time. Instead of checking pages randomly, the editor could inspect the highest-ranked candidates first and decide whether to refresh, expand, protect, prune, or monitor each page.

A wrong recommendation has two main costs. A false positive can waste editing time on a page that did not need attention. A false negative can miss a page with meaningful demand that continues to lose visibility or engagement. Because editor time is limited, I will care especially about the quality of the top part of the ranking, using a metric such as **precision@K**.

This is not just “train a model.” I will first define a transparent rule-based baseline, compare it with a simple learned method only if the data supports it, validate the ranking, and inspect the recommended pages by hand. Data or ML may help because several signals interact, and a fixed rule may not combine them well for every client or content type.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Load the starter CSV robustly in Colab, whether the notebook was opened
# from a cloned repo or directly from GitHub.

from pathlib import Path
import pandas as pd

possible_paths = [
    Path("data/raw/content_refresh_anonymized.csv"),
    Path("../../data/raw/content_refresh_anonymized.csv"),
    Path("/content/flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv"),
]

data_path = next((p for p in possible_paths if p.exists()), None)

if data_path is not None:
    df = pd.read_csv(data_path)
    source_used = str(data_path)
else:
    # Public fallback: load the starter CSV directly from the official repository.
    raw_url = (
        "https://raw.githubusercontent.com/"
        "flyrank-bih/flyrank-ml-internship-starter/main/"
        "data/raw/content_refresh_anonymized.csv"
    )
    df = pd.read_csv(raw_url)
    source_used = raw_url

print("Loaded from:", source_used)
print("Shape:", df.shape)
df.head()
```

```python
# Calculate real supporting numbers for the selected lane.

required = {
    "content_id", "client_id", "impressions_90d", "sessions_90d",
    "content_age_days", "trend_direction", "avg_position", "ctr"
}
missing = required.difference(df.columns)
if missing:
    raise ValueError(f"Expected columns are missing: {sorted(missing)}")

total_pages = df["content_id"].nunique()
total_clients = df["client_id"].nunique()

eligible = df[
    (df["impressions_90d"].fillna(0) > 0) &
    (df["content_age_days"].fillna(0) >= 90)
].drop_duplicates("content_id")

declining_with_demand = eligible[
    (eligible["trend_direction"].astype(str).str.lower() == "down") &
    (eligible["impressions_90d"].fillna(0) >= 100)
]

visible_low_ctr = eligible[
    (eligible["impressions_90d"].fillna(0) >= 500) &
    (eligible["avg_position"].fillna(0) > 0) &
    (eligible["avg_position"].fillna(0) <= 20) &
    (eligible["ctr"].fillna(0) < 0.5)
]

print(f"1. The starter dataset contains {len(df):,} rows, "
      f"{total_pages:,} unique pages, and {total_clients:,} clients.")
print(f"2. {len(eligible):,} pages have positive impressions and are at least "
      f"90 days old, so they have enough basic history for an initial review.")
print(f"3. {len(declining_with_demand):,} pages are declining with at least "
      f"100 impressions, while {len(visible_low_ctr):,} visible pages have "
      f"at least 500 impressions, positions 1–20, and CTR below 0.5%.")

supporting_numbers = {
    "rows": len(df),
    "unique_pages": total_pages,
    "clients": total_clients,
    "eligible_pages": len(eligible),
    "declining_with_demand": len(declining_with_demand),
    "visible_low_ctr": len(visible_low_ctr),
}
supporting_numbers
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# A small claim-safety check.
allowed_language = [
    "observed association",
    "directional pattern",
    "review candidate",
    "decision-support",
    "measured on the validation set",
]

not_allowed_without_stronger_evidence = [
    "refreshing caused recovery",
    "this is a Google ranking factor",
    "the recommendation guarantees traffic growth",
]

print("Language I can use:")
for phrase in allowed_language:
    print("-", phrase)

print("\nClaims I will not make from observational data:")
for phrase in not_allowed_without_stronger_evidence:
    print("-", phrase)
```

This project may identify **observed associations**, directional patterns, and pages that deserve human review based on measured search and engagement signals. It may support the decision of where an editor should look first, and it may show whether a learned ranking performs better than a simple baseline on a clearly defined validation set.

It cannot prove that refreshing a page caused a later recovery unless a proper experiment or causal design is used. It cannot claim to discover Google ranking factors, predict Google itself, or guarantee that a recommended edit will increase traffic. Low CTR may have several explanations, and a traffic drop may reflect seasonality, consolidation into another page, tracking availability, or normal noise. Therefore, recommendations will use cautious labels such as **review candidate**, **observed decline risk**, or **opportunity**, rather than guaranteed outcomes.
